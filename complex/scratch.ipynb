```julia
function lim(f::Function, c::Real; n::Int = 6, dir = "+")
    hs = [(1/10)^i for i in 1:n]
    if dir == "+"
        xs = c + hs
    else
        xs = c - hs
    end
    ys = map(f,xs)
    [xs ys]
end
```
