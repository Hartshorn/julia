```julia
using Gadfly
using Roots
```

```julia
function plotif(f, g, a, b)
    plot([f, x -> g(x) > 0.0 ? f(x) : NaN], a, b)
end
```

```julia
f(x) = x^3 - 2x - 1/2
plotif(f, f, -2, 2) # where f(x) > 0
```

```julia
plotif(f, D(f), -2, 2) # where the function is increasing (D(f) > 0)
```

```julia
plotif(f, D2(f), -2, 2) # where the function is concave up
```

```julia
f(x) = x^x
plotif(f, D(f), 0, 2)
```

```julia
f(x) = sqrt(abs(1 - x^2))
plotif(f, D(f), -2, 2)
```

```julia
f(x) = x^2 * exp(-x)
plotif(f, D2(f), 0, 10)
```

```julia
# Relationships
```

```julia
fp(x) = exp(x)
plotif(fp, fp, 0, 4)
```

```julia
plotif(fp, D(fp), 0, 4)
```

```julia
fpp(x) = x^2 - 2x
plotif(fpp, fpp, -1, 3)
```

```julia
fp(x) = abs(x)
plotif(fp, fp, -1, 1)
```

```julia
plotif(fp, D(fp), -1, 1)
```

```julia
fp(x) = (x^4 - x^2 + 2) / (x^4 - 2x^2 + 1)
plotif(fp, fp, -2, 2)
```

```julia
fp(x) = tan(x) - 3x/2
plotif(fp, D(fp), -pi/3, pi/3)
```

```julia
f(x) = 2sin(x) + cos(2x)
plot([f, D(f)], 0, 2pi)
```

```julia
# observe D(f) is zero at: 0.8, 1.5, 2.5, 5
#
# (three ways to say the same thing!)
#
# [fzero(D(f), x) for x in [0.8, 1.5, 2.5, 5]]
#
# map([0.8, 1.5, 2.5, 5]) do x
#     fzero(D(f), x)
# end
#
map(x -> fzero(D(f), x), [0.8, 1.5, 2.5, 5])
```

```julia
fp(x) = x^4 - 4x^2 + 1
plot(fp, -2, 2)
```

```julia
map(x -> fzero(fp, x), [-0.5, 0.5])
```

```julia
fp(x) = (2x - 1) / cbrt(x^2 - x - 2)^2
plot(fp, -1.5, 1.5)
```

```julia
fzero(fp, 0.5)
```
